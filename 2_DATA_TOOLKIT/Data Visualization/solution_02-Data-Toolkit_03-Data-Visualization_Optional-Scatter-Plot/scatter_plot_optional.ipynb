```python
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
import seaborn as sns
```

```python
# First, run the line below to download the dataset
!curl https://wagon-public-datasets.s3.amazonaws.com/02-Data-Toolkit/03-Data-Visualization/blobs.csv > blobs.csv
```

```python
data = pd.read_csv('blobs.csv')
data.head()
```

## 1. Size and transparency 
Increase the **visibility** of your scatter plot with **smaller** dots and a **lower opacity**.
Create a plot with **both** techniques used at the same time.

```python
sns.scatterplot(data=data, x='A', y='B', alpha=.05, s=25, markers='o');
```

## 2. Groups
Recreate this plot and use the `hue` option in seaborn scatter plot to **group by category**.


```python
sns.scatterplot(data=data, x='B', y='A', hue='Category');
```

## 3. Faceting and group highlight
Create a **grid** of scatter plots, with each group **highlighted**

```python
g = sns.FacetGrid(data=data, col='Category', hue='Category')
g.map(sns.scatterplot, "A", "B");
```

```python
sns.relplot(data=data, x='A', y='B', hue='Category', col='Category');
```
